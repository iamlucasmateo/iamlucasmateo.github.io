```python
# Chequeo que no haya NaN para graficar (podria rellenar sino)

data_grafico = {muestra_barrios[i]:list(barrio_props.loc[muestra_barrios[i]]) for i in range(len(muestra_barrios)) }
data = np.array(list(data_grafico.values()))
data
```

```python
# Horizontal bar chart, discrete distributions

# Armo variables para Matplotlib

cat_prop = ['Departamentos','PH','Casas']

# Armo el grafico

def horizontal_bar(data_grafico,cat_prop):

    labels = list(data_grafico.keys())
    data = np.array(list(data_grafico.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.get_cmap('RdYlGn')(np.array([0.1,0.35,1.5]))
    #(np.linspace(0.1, 1, data.shape[1]))

    fig, ax = plt.subplots(figsize=(12, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(cat_prop, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        ax.barh(labels, widths, left=starts, height=0.5,label=colname, color=color)
        xcenters = starts + widths / 2

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.3 else 'black'
        
        for y, (x, c) in enumerate(zip(xcenters, widths)):
            ax.text(x, y, str(int(c)), ha='center', va='center',color=text_color)
    
    ax.legend(ncol=len(cat_prop), bbox_to_anchor=(0, 1),loc='lower left', fontsize=13)
    ax.set_title('TIPOS DE PROPIEDADES POR BARRIOS')
    ttl = ax.title
    ttl.set_position([.5, 1.15])
    ttl.set_fontsize(17)
    ttl.set_fontweight('bold')


horizontal_bar(data_grafico,cat_prop)

```

De este modo se puede apreciar que existe una enorme variabilidad en el tipo de propiedad que se vende en cada barrio.
En Palermo mas del 90% de las propiedades son departamentos, mientras que en Villa Devoto el numero se reduce a cerca del 50%.

### Barrios que quieren ser otros

Basado en información de mercado, desde el portal Properati han señalado que las inmobiliarias suelen publicar las propiedades ubicadas en barrios que no necesariamente se corresponden con la ubicación geográfica real. Según mencionan en la nota ([blog)](http://blog.properati.com.ar/barrios-que-quieren-ser-otros/), esto corresponde a una lógica comercial en el corredor norte y a usos y costumbres en zona oeste.

Analicemos qué sucede para nuestro dataset.

Dada la estructura del atributo `place_with_parent_names`, en primer lugar vamos a extraer el campo asociado al barrio publicado para poder comparar con el que surge de la cartografía oficial. Esto lo vamos a agregar en una nueva columna llamada `barrio_publicado`. 

De este modo, lo que vamos a analizar es como se relacionan las columnas `barrio` y `barrio publicado`. Si efectivamente hay barrios que quieren ser otros, esperaríamos ver propiedades ubicadas en geográficamente en un barrio distinto al que dice la publicación. Por ejemplo, si una propiedad está ubicada (dadas sus coordenadas) en Villa Crespo pero en la publicación dice Palermo, entonces decimos que hay indicios que Villa Crespo quiere ser Palermo.

Viendo la estructura de este atributo, podemos notar que se compone por una sequencia de lugares separados por `|`. Por ejemplo, en `|Argentina|Capital Federal|Palermo|` vemos que primero está el país, luego la ciudad, luego el barrio y en algunos casos hay una jerarquía más, relacionado con zonas dentro del barrio (como Palermo Soho dentro de Palermo). 

De este modo, una estrategia posible para resolver esto es separar las jerarquías por `|` y extraer el elemento asociado al barrio. Por ejemplo, para `|Argentina|Capital Federal|Palermo|` queda `['', 'Argentina', 'Capital Federal', 'Palermo', '']`. Extrayendo el tercer elemento obtendremos el barrio independientemente de si hay lugares luego en la jerarquía. Aquellos que no tengan barrio no serán considerados puesto que estarían ubicados en Capital Federal y no sería informativo.

Lo pasamos a mayúsculas y quitamos las tildes para poder comparar con el barrio oficial.

```python
df_filtered['place_with_parent_names'].apply(lambda x: len(x.split("|"))).unique()
```

Como algunos tienen 4 elementos, debemos quitarlos.

```python
split_place_filter = df_filtered['place_with_parent_names'].apply(lambda x: len(x.split("|"))!=4)
df_filtered = df_filtered.loc[split_place_filter]
```

```python
df_filtered['barrio_publicado'] = df_filtered['place_with_parent_names'].apply(lambda x: x.split("|")[3].upper())
df_filtered['barrio_publicado'] = df_filtered['barrio_publicado'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

Comparemos entonces el barrio publicado con el real. Nos vamos a quedar con aquellas propiedades donde nos coincidan para luego hacer el conteo por barrio.

```python
barrios_match = (df_filtered['barrio_publicado'] != df_filtered['barrio'])
df_filtered_barrio_not_match = df_filtered.loc[barrios_match]
```

Con una tabla de doble entrada podemos identificar para cada barrio cuál es al que más quiere parecerse.

```python
table = pd.crosstab(df_filtered_barrio_not_match.barrio, df_filtered_barrio_not_match.barrio_publicado)
table.idxmax(axis=1)
```

Podemos ver entonces que los barrios tienden a querer parecerse a barrios con precio con metro cuadrado más caro. Por ejemplo Villa Crespo con Palermo, Constitución con San Telmo y Villa Urquiza con Belgrano. 

Por otro lado, vemos también que los barrios suelen querer parecerse a sus barrios vecinos.

# Análisis: ¿cuánto vale estar cerca del subte?

<br><div style="text-align: justify">Me interesa analizar la relacion entre la cercania a una boca de subte y el valor de una propiedad. Es un lugar común la idea de que una propiedad cercana a una boca de un subte tiene, ceteris paribus, un valor mayor que una lejana. En nuestro país, dicho lugar común está inclusive consagrado por ley 23.514 de 1987, que establece que las personas que posean propiedades en un radio de 400 metros de una boca de subte deberán pagar una "contribución de mejoras" (es decir, un impuesto ad hoc) cuando se habilite un tramo nuevo de la red. <br>

```python
# Importo lat-lon de bocas de subte (Fuente: BA Data)
df_subte_lugar = pd.read_csv('bocas-de-subte.csv')
print(df_subte_lugar.shape)

# Muestra para chusmear
df_subte_lugar.iloc[0:7,0:5]
```

### Buscando la distancia a la boca de subte más cercana

La celda que aparece a continuación es la que más tarda en ejecutar de todo el notebook (unos dos minutos en mi compu). Esto es porque el algoritmo no está muy optimizado, pero como lo corro una sola vez, me resulta aceptable. Me tomaría más tiempo crear un algoritmo más eficiente de lo que me toma esperar esos dos minutos, me parece. Por lo menos a mí.


```python
# Copia del dataframe para este analisis, reindexo para trabajar mejor despues
df_subte_pr = df_filtered.copy()
df_subte_pr.index = range(len(df_subte_pr))

# Vamos a hallar la minima distancia a una boca de subte

# Primero defino distancia euclideana (la curvatura de la tierra no me cambia para esta escala)

def distance(lon1,lat1,lon2,lat2):
    # Longitud aprox de meridianos / paralelos en metros
    deglen = 110500
    delta_x = lat1 - lat2
    delta_y = (lon1 - lon2)*np.cos((lat1 + lat2) / 2)
    return deglen*((delta_x)**2 + (delta_y)**2)**(0.5)

# Calculo distancia de cada propiedad a cada boca de subte 

for i in range(len(df_subte_pr)):
    
    # Distancia con la primera boca 
    lon1 = df_subte_pr.loc[i,'lon']
    lat1 = df_subte_pr.loc[i,'lat']
    lon2 = df_subte_lugar.iloc[0,0]
    lat2 = df_subte_lugar.iloc[0,1]
    estacion_id = 1
    min_dist = distance(lon1,lat1,lon2,lat2)
    
    # Esto de abajo es un control para ir viendo como avanza, cada mil muestras, de ansioso nomas
    
    #if abs(i % 1000 - 0) < 0.00001:
        #print(i)
    
    # Calculo distancia a cada boca y me voy quedando con el minimo
    for j in range(1,len(df_subte_lugar)):
        lon2 = df_subte_lugar.iloc[j,0]
        lat2 = df_subte_lugar.iloc[j,1]
        dist_aux = distance(lon1,lat1,lon2,lat2)
        if dist_aux < min_dist:
            min_dist = dist_aux
            estacion_id = df_subte_lugar.loc[j,'id']
    
    # Asigno el minimo hallado a dos columnas nuevas
    df_subte_pr.loc[i,'estacion'] = estacion_id
    df_subte_pr.loc[i,'min_dist'] = min_dist
```

```python
# Le pongo unos nombre mas lindos
df_subte_pr.rename(columns={'price_usd_per_m2':'Precio M2', 'min_dist': 'Distancia Subte'},inplace=True)

# Veo una muestra
df_subte_pr.loc[:,['created_on','barrio','Precio M2','Distancia Subte','lat-lon','estacion']]
```

En conclusión, la relación entre el precio del m2 y la distancia a una boca del subte está muy lejos de ser unívoca. La fuerza de esta relación puede estudiarse más formalmente a través de tests de significación, pero informalmente podríamos indagar en diferentes razones por las cuales el mito no se cumple. La primera y más obvia es la posible baja representatividad de la muestra (lo cual haría todo el análisis un poco superfluo).

Fuera de esto, si aceptamos la representatividad de la muestra, las correlaciones positivas son fáciles de entender; son la sabiduría convencional. Con respecto a las correlaciones negativas, uno podría sentirse tentado de asignarlas al azar o a sesgos muestrales; sin embargo, también podrían estar señalando causalidades más sutiles en la distribución geográfica de los precios. Dos variables negativamente correlacionadas pueden no tener una relación causal directa y sin embargo reflejar una tercera variable que las condiciona a ambas. Es decir, no necesariamente las propiedades más cercanas al subte son menos atractivas por el hecho mismo de estar más cerca del subte; más bien, estar cerca del subte podría implicar otras cualidades menos deseables. A modo ejemplo -y entrando ya descaradamente en el terreno de la especulación-, en un barrio como Barracas la cercanía con el subte implica estar lejos de la Gral Paz, que sería una solución de transporte más deseable para esa zona. En Villa Urquiza, estar cerca del subte es estar cerca de la zona más céntrica y comercial del barrio, lo que desde el punto de vista residencial no es necesariamente lo más deseable. Sería interesante comparar con los precios de los locales comerciales.

Las dos tendencias podrían estar operando en el caso de los barrios sin correlación; por ejemplo, en un barrio completamente atravesado por el subte como Balvanera, la cercanía a una boca puede implicar cierta comodidad, pero también implica la molestia de vivir en una zona de mucho tránsito permanente. Si estas dos tendencias cotradictorias se compensan, la correlación entre valor y distancia será nula. En Villa Pueyrredón, estar cerca del subte puede ser una ventaja a la hora de ir al centro, pero a la vez implica alejarse de la Gral Paz y sus opciones de transporte.

Naturalmente, el tema amerita un análisis más profundo. Por lo pronto me siento satisfecho con la metodología, que podría ser ampliada a otras variables mapeadas por el Gobierno de la Ciudad (espacios verdes, sitios posibles de anegamiento, comisarías).
