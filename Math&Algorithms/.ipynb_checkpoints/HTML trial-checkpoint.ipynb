```python
df = sns.load_dataset('titanic')
df = df[['sex','class','alive']]
df
```

```python
df.drop(columns=['x','y'],inplace=True)
```

We'll generate artificial coordinates to be able to see our categorical distribution more clearly

```python
females = df[df['sex'] == 'female'].shape[0]
males = df[df['sex'] == 'male'].shape[0]
third = df[df['class'] == 'Third'].shape[0]
second = df[df['class'] == 'Second'].shape[0]
first = df[df['class'] == 'First'].shape[0]

x_fem = np.random.random(females)*3
x_male = np.random.random(males)*3 + 3
y_third = np.random.random(third)*2
y_sec = np.random.random(second)*2 + 2
y_first = np.random.random(first)*2 + 4

df['x'] = 0
df.sort_values(by='sex',inplace=True)
x_coord = np.concatenate((x_fem,x_male))
df['x'] = x_coord

df['y'] = 0
df.sort_values(by='class',inplace=True)
y_coord = np.concatenate((y_first,y_sec,y_third))
df['y'] = y_coord

df['dummy'] = np.where(df.alive == 'yes',1,0)

df
```

```python
# Create plot, etc
fig, ax = plt.subplots(figsize=(10,10))

label = list(df['dummy'])
colors = ['red','blue']

ax.scatter(df['x'], df['y'],c=label, cmap=matplotlib.colors.ListedColormap(colors),edgecolors='#333333')

plt.ylim(-0.2,6.2)
plt.xlim(-0.2,6.2)

# Background color
ax.set_facecolor((0.97,0.97,0.97))

ax.grid(linewidth=2,color='#555555')
ax.set_yticks([0, 2, 4,6], minor=False)
ax.set_xticks([0, 3, 6], minor=False)
ax.yaxis.grid(True, which='major')
ax.yaxis.grid(True, which='minor')

font_dictx = {'fontsize': 20,'fontweight': 'bold','verticalalignment': 'baseline','horizontalalignment': 'left'}

font_dicty = {'fontsize': 20,'fontweight': 'bold','verticalalignment': 'baseline','horizontalalignment': 'right'}

font_xaxis = {'fontsize': 30,'fontweight': 'bold','verticalalignment': 'bottom','horizontalalignment': 'right'}

ax.set_xticklabels(['Male','Female'],fontdict=font_dictx)
ax.set_yticklabels(['First','Second','Third'],fontdict=font_dictx)

plt.text(3,-0.8,'Sex',fontdict=font_xaxis)
plt.text(-0.4,3,'Class',fontdict=font_xaxis,rotation=90)

ax.clip_box = None

# a = ax.get_xgridlines()
# b = a[4]
# b.set_color('black')
# b.set_linewidth(10)

alive_leg = mpatches.Patch(color='red',label='Alive')
dead_leg = mpatches.Patch(color='blue',label='Dead')
plt.legend(handles=[alive_leg,dead_leg],fontsize='large')

plt.show()
```

```python
# Create plot, etc
fig, ax = plt.subplots(figsize=(10,10))

label = list(df['dummy'])
colors = ['red','blue']

ax.scatter(df['x'], df['y'],c=label, cmap=matplotlib.colors.ListedColormap(colors),edgecolors='#333333')

plt.ylim(-0.2,6.2)
plt.xlim(-0.2,6.2)

# Background color
ax.set_facecolor((0.97,0.97,0.97))
plt.axvspan(0, 3, facecolor='1.0', alpha=0.8)

ax.grid(linewidth=2,color='#555555')
ax.set_yticks([0, 2, 4,6], minor=False)
ax.set_xticks([0, 3, 6], minor=False)
ax.yaxis.grid(True, which='major')
ax.yaxis.grid(True, which='minor')


ax.set_xticklabels(['Male','Female'],fontdict=font_dictx)
ax.set_yticklabels(['First','Second','Third'],fontdict=font_dictx)

plt.text(3,-0.8,'Sex',fontdict=font_xaxis)
plt.text(-0.4,3,'Class',fontdict=font_xaxis,rotation=90)


a = ax.get_xgridlines()
b = a[1]
b.set_color('black')
b.set_linewidth(10)

alive_leg = mpatches.Patch(color='red',label='Alive')
dead_leg = mpatches.Patch(color='blue',label='Dead')
plt.legend(handles=[alive_leg,dead_leg],fontsize='large')

plt.show()
```
